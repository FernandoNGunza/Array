```python
x=['Manga', 'Laranja', 'Limão', 'Banana']
y=[3,5,7,9]

plt.bar(x,y)

plt.xlabel('Frutas')
plt.ylabel('Quantidade')
plt.title('Quantidade de Frutas')
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[11,22,33,4,35,66,47]


plt.scatter(x,y, label='Pontos', color='b', marker='o')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Dispersão')
plt.show()
```

```python
titanic=sns.load_dataset('titanic')
print(titanic)


df_sexo=titanic.groupby('sex')['survived'].sum().reset_index

plt.figure(figsize=(6, 4))


sns.barplot(x='sex', y='survived', data=titanic)

```

```python
plt.figure(figsize=(6, 4))
sns.histplot(titanic['age'], bins=10, kde=True)
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.title('Distribuição de Idades')
plt.show()
```

```python
voos=sns.load_dataset('flights')


plt.figure(figsize=(8, 6))
sns.heatmap(voos.pivot(index='year', columns='month', values='passengers'), annot=True, fmt='d', cmap='YlGnBu')
plt.xlabel('Ano')
plt.ylabel('Passageiros')
plt.title('Voo de Passageiros ao Longo do Tempo')
plt.show()
```
